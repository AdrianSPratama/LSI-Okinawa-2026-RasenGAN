```python
def bil_to_hex(x):
    x = int(x)  # <--- Tambahkan baris ini
    if x < 0:
        x = (1 << 16) + x
    return hex(x)
```

## BIAS

### Ekstrak bias from bin

```python
bias_conv_4 = np.fromfile('weights_bias_4x4.bin', dtype=np.int16)

bias_conv1_8 = np.fromfile('weights_bias1_8.bin', dtype=np.int16)
bias_conv2_8 = np.fromfile('weights_bias2_8.bin', dtype=np.int16)

bias_conv1_16 = np.fromfile('weights_bias1_16.bin', dtype=np.int16)
bias_conv2_16 = np.fromfile('weights_bias2_16.bin', dtype=np.int16)

bias_conv1_32 = np.fromfile('weights_bias1_32.bin', dtype=np.int16)
bias_conv2_32 = np.fromfile('weights_bias2_32.bin', dtype=np.int16)

bias_conv1_64 = np.fromfile('weights_bias1_64.bin', dtype=np.int16)
bias_conv2_64 = np.fromfile('weights_bias2_64.bin', dtype=np.int16)

bias_conv1_128 = np.fromfile('weights_bias1_128.bin', dtype=np.int16)
bias_conv2_128 = np.fromfile('weights_bias2_128.bin', dtype=np.int16)
```

```python
all_biases = np.concatenate((bias_conv_4, bias_conv1_8, bias_conv2_8, bias_conv1_16, bias_conv2_16, bias_conv1_32, bias_conv2_32, bias_conv1_64, bias_conv2_64, bias_conv1_128, bias_conv2_128))
```

### Save to file .mem

```python
with open('biases.mem', mode='w', newline='') as file:
    writer = csv.writer(file)


    for i in range(len(all_biases)):
        # tulis ke file csv
        try :
            writer.writerow([f'{bil_to_hex(all_biases[i])[2:]}'])

        except :
            print(f'Error at index {i} with value {all_biases[i]}')
```
