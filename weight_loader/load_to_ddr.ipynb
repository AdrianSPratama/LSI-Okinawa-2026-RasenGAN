## Memberikan 7 zero padding ke weight untuk setiap 9 data (32 byte format)

```python
def zero_pad_weights(weights):
    i = 0
    res = np.zeros(len(weights) + 7*len(weights)//9).flatten()

    counter = 0
    for w in weights:

        if counter == 8:
            i += 7
            counter = 0
        res[i] = w
        counter += 1
        i += 1
    return res


print((weights_conv_4.shape))
x = zero_pad_weights(weights_conv_4)
print((x[:20]))
```
